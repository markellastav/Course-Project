```python
#data description

#load data
import numpy as np
file = open(input()) #input file, .tsv format
biomasses = [] #create empty list of lists

#read file
for line in file: #loop
    data = line.split("\t") #split each line on tabs
    if line[0] == "#": #grab header
        header = data
    else:
        out = []
        for i in data[4:18]: #include select numeric data
            out.append(float(i)) #transform to floats
        biomasses.append(out) #add to list
        
new = np.array(biomasses) #transform to array
transpose = np.transpose(new) #flip rows and columns

snails = transpose[1] #create variable for each relevant species biomass
JFC = transpose[3]
AFC = transpose[5]
mussel = transpose[8]
mud = transpose[10]
sesarma = transpose[12]

#mean
mean_snails = np.mean(snails) #calculate the mean of species biomass data
mean_JFC = np.mean(JFC)
mean_AFC = np.mean(AFC)
mean_mussel = np.mean(mussel)
mean_mud = np.mean(mud)
mean_sesarma = np.mean(sesarma)

#median
med_snails = np.median(snails) #calculate the median of species biomass data
med_JFC = np.median(JFC)
med_AFC = np.median(AFC)
med_mussel = np.median(mussel)
med_mud = np.median(mud)
med_sesarma = np.median(sesarma)

#std dev
st_snails = np.std(snails) #calculate the standard deviation of species biomass data
st_JFC = np.std(JFC)
st_AFC = np.std(AFC)
st_mussel = np.std(mussel)
st_mud = np.std(mud)
st_sesarma = np.std(sesarma)

#print results
print("Snail Mean: {:.3f},".format(mean_snails), "Snail Median: {:.3f},".format(med_snails), "Snail Standard Deviation: {:.3f}".format(st_snails))
print("JFC Mean: {:.3f},".format(mean_JFC), "JFC Median: {:.3f},".format(med_JFC), "JFC Standard Deviation: {:.3f}".format(st_JFC))
print("AFC Mean: {:.3f},".format(mean_AFC), "AFC Median: {:.3f},".format(med_AFC), "AFC Standard Deviation: {:.3f}".format(st_AFC))
print("Mussel Mean: {:.3f},".format(mean_mussel), "Snail Median: {:.3f},".format(med_mussel), "Snail Standard Deviation: {:.3f}".format(st_mussel))
print("Mud Crab Mean: {:.3f},".format(mean_mud), "Mud Crab Median: {:.3f},".format(med_mud), "Mud Crab Standard Deviation: {:.3f}".format(st_mud))
print("Sesarma Mean: {:.3f},".format(mean_sesarma), "Sesarma Median: {:.3f},".format(med_sesarma), "Sesarma Standard Deviation: {:.3f}".format(st_sesarma))

#histograms
import matplotlib.pyplot as plt #import

fig, ax = plt.subplots() #set up on xy axis
plt.hist(snails) #plot histogram
plt.title("Snail Biomass") #title graph
label_X = "Biomasses" #label x axis
label_Y = "Frequency" #label y axis
ax.set_xlabel(label_X)
ax.set_ylabel(label_Y)
plt.show()

#repeat for each species
fig, ax = plt.subplots()
plt.hist(JFC)
plt.title("JFC Biomass") #title graph
label_X = "Biomasses" #label x axis
label_Y = "Frequency" #label y axis
ax.set_xlabel(label_X)
ax.set_ylabel(label_Y)
plt.show()

fig, ax = plt.subplots()
plt.hist(AFC)
plt.title("AFC Biomass") #title graph
label_X = "Biomasses" #label x axis
label_Y = "Frequency" #label y axis
ax.set_xlabel(label_X)
ax.set_ylabel(label_Y)
plt.show()

fig, ax = plt.subplots()
plt.hist(mussel)
plt.title("Mussel Biomass") #title graph
label_X = "Biomasses" #label x axis
label_Y = "Frequency" #label y axis
ax.set_xlabel(label_X)
ax.set_ylabel(label_Y)
plt.show()

fig, ax = plt.subplots()
plt.hist(mud)
plt.title("Mud Crab Biomass") #title graph
label_X = "Biomasses" #label x axis
label_Y = "Frequency" #label y axis
ax.set_xlabel(label_X)
ax.set_ylabel(label_Y)
plt.show()

fig, ax = plt.subplots()
plt.hist(sesarma)
plt.title("Sesarma Biomass") #title graph
label_X = "Biomasses" #label x axis
label_Y = "Frequency" #label y axis
ax.set_xlabel(label_X)
ax.set_ylabel(label_Y)
plt.show()
```

```python
#dimensionality reduction

#load data
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats #import scipy
from sklearn.decomposition import PCA
file = open(input()) #input file, .tsv format
quantitative = [] #create empty list of lists

#read file
for line in file: #loop
    data = line.strip().split("\t") #split each line on tabs
    if line[0] == "#": #grab header
        header = data
    else:
        out = []
        for i in data[4:18]: #include select numeric data
            out.append(float(i)) #transform to floats
        quantitative.append(out) #add to list

data = np.array(quantitative)
raw_data = data #save original data

new_data = np.delete(data, 0, axis=1)
new_data = np.delete(new_data, 1, axis=1)
new_data = np.delete(new_data, 2, axis=1)
new_data = np.delete(new_data, 3, axis=1)
new_data = np.delete(new_data, 3, axis=1)
new_data = np.delete(new_data, 4, axis=1)
new_data = np.delete(new_data, 5, axis=1)
new_data = np.delete(new_data, 6, axis=1)

mean = np.mean(new_data, axis = 0)
center = new_data - mean

covariance = np.cov(center, rowvar=False)

eigenvalues, eigenvectors = np.linalg.eigh(covariance)

index = eigenvalues.argsort()[::-1]
components = eigenvectors[:, index]

axis = components[:, 0]
plt.figure(figsize = (8, 6))
plt.bar(range(1, len(axis) + 1), axis)
plt.xlabel("Species Biomass")
plt.ylabel("Components")
plt.title("Weighted Principal Components")
plt.xticks(range(1, len(axis) + 1), ["Snails", "JFC", "AFC", "Mussels", "Mud Crabs", "Sesarma Crabs"])

#initialize PCA object
pca = PCA(n_components=6)

#fit PCA model to data
#pca.fit(data)
principal_components = pca.fit_transform(new_data)

#plot PCA

#generate plot
fig, ax = plt.subplots() #create figure w/ single axes

PCX = 1 #need PCA 1 and 2
PCY = 2

ax.scatter(principal_components[:71, PCX], principal_components[:71, PCY], c= "orange") #grazed, border
ax.scatter(principal_components[72:144, PCX], principal_components[72:144, PCY], c = "blue") #grazed, platform
ax.scatter(principal_components[145:216, PCX], principal_components[145:216, PCY], c = "red") #ungrazed, border
ax.scatter(principal_components[216:, PCX], principal_components[216:, PCY], c = "green") #ungrazed, platform

#set axis lables
plt.title("PC1 vs. PC2") #title graph
label_X = "PC1 (" + str(np.round(pca.explained_variance_ratio_[PCX]*100, decimals = 2)) + "%)" #label x axis
label_Y = "PC2 (" + str(np.round(pca.explained_variance_ratio_[PCY]*100, decimals = 2)) + "%)" #label y axis
ax.set_xlabel(label_X)
ax.set_ylabel(label_Y)
```
